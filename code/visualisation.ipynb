```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits
%matplotlib inline
```

```python
clean_CMO = "../dataset/clean_CMO.csv"
clean_monthly = "../dataset/clean_monthly.csv"
```

```python
CMO = pd.read_csv(clean_CMO, sep='\t')
monthly = pd.read_csv(clean_monthly, sep='\t')
CMO = CMO.drop('Unnamed: 0', axis=1)
monthly = monthly.drop('Unnamed: 0', axis=1)
```

```python
#CMO.head(1)
monthly.head(1)
monthly.describe()
```

```python
CMO['commodity'].value_counts().plot(kind='bar')
plt.title('Commodities')
plt.xlabel('Commodities')
plt.ylabel('Count')
plt.savefig('../graphs/Commodities.png', bbox_inches='tight', dpi=300)
sns.despine
```

```python
CMO['Type'].value_counts().plot(kind='bar')
plt.title('Type of crops')
plt.xlabel('Type')
plt.ylabel('Count')
plt.savefig('../graphs/Type of crops.png', bbox_inches='tight', dpi=300)
sns.despine
```

```python
plt.figure(figsize=(8,5))
plt.scatter(monthly.year, monthly.modal_price)
plt.title("modal_price vs year")
plt.savefig('../graphs/modal price vs year.png', bbox_inches='tight', dpi=300)
sns.despine
```

```python
plt.figure(figsize=(8,5))
plt.scatter(monthly.Month, monthly.modal_price)
plt.title("modal_price vs month")
plt.savefig('../graphs/modal price vs month.png', bbox_inches='tight', dpi=300)
sns.despine
```

```python
plt.figure(figsize=(8,5))
plt.scatter(monthly.arrivals_in_qtl, monthly.modal_price)
plt.title("modal_price vs arrivals(qtl)")
plt.xlabel('arrivals(qtl)')
plt.ylabel('modal price')
plt.savefig('../graphs/modal price vs arrivals(qtl).png', bbox_inches='tight', dpi=300)
sns.despine
```

```python
plt.figure(figsize=(8,5))
plt.scatter(monthly.max_price, monthly.modal_price)
plt.title("modal_price vs max price")
plt.xlabel('max price')
plt.ylabel('modal price')
plt.savefig('../graphs/modal price vs max price.png', bbox_inches='tight', dpi=300)
sns.despine
```

```python
plt.figure(figsize=(8,5))
plt.scatter(monthly.year, monthly.arrivals_in_qtl)
plt.title("arrivals(qtl) vs year")
plt.xlabel('year')
plt.ylabel('arrivals(qtl)')
plt.savefig('../graphs/arrivals(qtl) vs year.png', bbox_inches='tight', dpi=300)
sns.despine
```

```python
plt.figure(figsize=(8,5))
plt.scatter(monthly.commodity, monthly.arrivals_in_qtl)
plt.title("arrivals(qtl) vs commodity")
plt.xlabel('commodity')
plt.ylabel('arrivals(qtl)')
plt.savefig('../graphs/arrivals(qtl) vs commodity.png', bbox_inches='tight', dpi=300)
sns.despine
```
